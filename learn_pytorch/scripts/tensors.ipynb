```python
import torch
```

```python
torch.cuda.is_available()
```

```python
x = torch.arange(12, dtype=torch.float32)
x
```

```python
# total number of elements in the tensor
print(x.numel())

# tensor shape
print(x.shape)
```

```python
# change the shape of the tensor to (3,4)
X = x.reshape(-1, 4)
X
```

```python
# initiate a tensor of zeros with shape (2,3,4)
# this means 2 matrices of shape (3,4), where each matrix has 3 rows and 4 columns
Z = torch.zeros((2, 3, 4))
print(Z)

One = torch.ones((2, 3, 4))
print(One)

rand = torch.rand((2, 3, 4))
print(rand)
```

```python
# construct tensor from data
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
tensor_from_data = torch.tensor(data)
print(tensor_from_data)
```

```python
# indexing and slicing of tensors
print(tensor_from_data[-1]) # last row

print(tensor_from_data[1:3]) # rows from index 1 to 2

#writing elements to tensor
tensor_from_data[0, 0] = 99
print(tensor_from_data)

# assigning multiple elements
tensor_from_data[0:2, :] = 42 # all columns of first two rows
print(tensor_from_data)
```

```python
# Operations on tensors

# exp
print(torch.exp(tensor_from_data))
```

```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])
print(x + y, x - y, x * y, x / y, x ** y)
```

```python
X = torch.arange(12, dtype=torch.float32).reshape((3,4))
print(X)
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print(Y)
op1 = torch.cat((X, Y), dim=0) # concatenate along rows
print(op1)
op2 = torch.cat((X, Y), dim=1) # concatenate along columns
print(op2)
```

```python
# Broadcasting

a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
print(a)
print(b)
print(a + b)
```
